```python
#se importan las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
#se lee el archivo csv
data=pd.read_csv('real_estate.csv')
```

```python
#se verifica el contenido del archivo
data
```

```python
data.describe()
```

```python
# Busco predecir el precio de las propiedades segun su tamaño
#definir variable dependiente e independiente

#variable dependiente
y=data['price']
#variable independiente
x1=data['size']
```

### Exploración de datos

```python
plt.scatter(x1,y, c='m')
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```

### Regresión

```python
x=sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

### Sustitución de valores en la ecuación

```python
plt.scatter(x1, y, c='m')
yhat=101900+223.1787* x1  #los valores que se sustituyen son del coef const y size
fig =plt.plot(x1, yhat, lw=4, c='c', label='regresion lineal')
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```
